```
dicwt = {}
def import_inventory():
    a = open("products.txt", "r")
    b = a.read()
    a.close()
    global dicwt
    z = b.split("-")
    for i in range(len(z)):
        ind = z[i].find("_")
        dicwt[z[i][0:ind].lower()] = z[i][ind+1:]
    c = open("products.txt", "w")
    c.write(str(dicwt))
    c.close()
    return b


def sell_product(products):
    products = input()
    t = products.split("-")
    fin = ""
    global dicwt
    for i in range(len(t)):
        sf = t[i].find("_")
        if eval(dicwt[t[i][0:sf].lower()]) >= int(t[i][sf+1:]):
            dicwt[t[i][0:sf]] = int(dicwt[t[i][0:sf]]) - int(t[i][sf+1:])
            fin += f"{t[i][sf + 1:]} {t[i][0:sf]} sold." + "\n"
        else:
            fin += f"There is not enough {t[i][0:sf]} in inventory." + "\n"
    c = open("products.txt", "w")
    c.write(str(dicwt))
    c.close()
    print(fin)


def show_remaining(products):
    a = open("products.txt")
    b = a.read()
    a.close()
    if b == "":
        print("Inventory is empty.")
    else:
        print(b)
kas = import_inventory()
print(kas)
```

```

products = import_inventory()
decision = ""
while decision != "3":
  print("*---------------------------")
  print("[1] Sell products")
  print("[2] Show remaining inventory")
  print("[3] Terminate")

  decision = input("Please enter your decision: ")
  print("____")
  if decision == "1":
    sell_product(products)
  elif decision == "2":
    show_remaining(products)
  elif decision == "3":
    print("Terminating...")
  else:
    print("Invalid input!")
```
